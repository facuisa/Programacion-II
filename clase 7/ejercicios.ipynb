1. Ejercicio de Listas:
o Crear una lista de 10 números enteros. Escribir una función que
devuelva la suma de todos los números en la lista.



```python
# Creamos una lista de ejemplo con 10 números enteros
lista_numeros = [4, 8, -3, 7, 12, -1, 6, 9, -5, 10]
def suma_lista(numeros):
    suma = 0  # Inicializamos la variable para almacenar la suma
    for numero in numeros:  # Recorremos la lista con un bucle for
        suma += numero  # Sumamos cada número a la variable 'suma'
    return suma  # Devolvemos el resultado final
# Llamamos a la función y almacenamos el resultado
resultado = suma_lista(lista_numeros)
# Imprimimos el resultado
print(f"La suma de los números en la lista es: {resultado}")
```

2. Ejercicio de Tuplas:
o Definir una tupla con 5 nombres de ciudades. Escribir una función
que imprima cada ciudad en una línea separada.


```python
# Definimos una tupla con 5 nombres de ciudades
ciudades = ("Buenos Aires", "Madrid", "Tokio", "París", "Sídney")
# Función que recibe una tupla e imprime cada ciudad en una línea separada
def imprimir_ciudades(ciudades):
    for ciudad in ciudades:  # Recorremos cada ciudad en la tupla
        print(ciudad)  # Imprimimos la ciudad
# Llamamos a la función para imprimir las ciudades
imprimir_ciudades(ciudades)
```

4. Ejercicio de Control de Flujo:
o Escribir un programa que pida al usuario un número y use un
bucle for para imprimir todos los números desde 1 hasta ese
número. Usar break para salir del bucle si se ingresa un número
negativo.


```python
# Solicitar al usuario un número
numero = int(input("Por favor, ingresa un número: "))
# Verificar si el número es negativo
if numero < 0:
    print("Número negativo ingresado. Saliendo del bucle.")
else:
    # Usar un bucle for para imprimir los números desde 1 hasta el número ingresado
    for i in range(1, numero + 1):
        print(i)
```

```python
numero = int(input("Por favor, ingresa un número: "))
for i in range(1, numero + 1):
    if i < 0:
        break
    else:
        print(i)
```

5. Ejercicio de Números Pares:
o Crear una función que reciba una lista de números y devuelva una
nueva lista que contenga solo los números pares.
Utilizar continue para omitir los números impares.


```python
lista2 = [10, 15, 20, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80]
def numeros_pares(lista):
    pares = []
    for i in lista:
        if i  % 2 == 0:
            pares.append(i)
        else:
            continue
    return pares
```

```python
mostrar_pares = numeros_pares (lista2)
print(mostrar_pares)
```
