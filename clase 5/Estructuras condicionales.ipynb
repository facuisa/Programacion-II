Estructura de control condicional

.if
.elif
.else

Las condicionales son estructuras fundamentales en programacion ya que permiten tomar decisiones y ejecutar
diferentes acciones en funcion de una condicion evaluada. Estas estructuras son clave para crear logica y 
controlar el flujo del programa.


```python
#ejemplo de utilizacion de if else
if condicion:
    print("esto es lo que hay que hacer")
else:
    print("esto es lo que no hay que hacer")
```

```python
edad = 10
if edad >= 18:
    print("puede acceder por ser mayor de edad")
else:
    print("no puede acceder por ser menor de edad")
```

```python
contraseña = "facu123"
contraseña_ingresada = "facu123"

if contraseña == contraseña_ingresada:
    print("puede ingresar al sistema")
else:
    print("contraseña incorrecta")
```

```python
sueldo = 80000
if sueldo <= 203000:
    print("el ingreso mensual es inferior a la canasta basica")
elif sueldo > 203000 and sueldo < 70000:
    print("el ingreso mensual es equivalente a la clase media baja")
elif sueldo >= 700000 and sueldo < 1200000:
    print("el ingreso mensual es de clase media alta")
elif sueldo  >= 1200000 and sueldo < 2500000:
    print("el ingreso mensual es de clase alta")
else:
    print("el ingreso mensual es de clase muy alta")
```
