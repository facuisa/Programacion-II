```python
def mi_funcion (*args, **kwargs):   #la cantidad de asteriscos es lo que define la funcion mas que el nombre
    print(f'args:{args}, kwargs {kwargs}')
    
mi_funcion()
```

```python
mi_funcion(29, 'facundo', True, 'Rio grande', vehiculo = 'coche' )

mi_funcion(vehiculo = 'coche')
```

```python
def sumar (*tupla1):    #significa que puedo usar muchos argumentos y que se pueden ir sumando 
    total = 0
    for i in tupla1:
        total += i
    return total 

#llamada a la funcion con argumentos posicionnales 
resultado = sumar (1, 2, 3, 4)
print(resultado)
```

```python
resultado2 = sumar (30, 100, 90, 909, 800, 121, 434)
```

```python
def mostrar_informacion (**kwargs):
    for clave, valor in kwargs.items():
        print(clave + ": " + valor)
        
#llamada a la funcion con argumentos clave-valor
mostrar_informacion (nombre='facundo', edad='29', musica_favorita='jazz, hip-hop, rock')

```

```python
numero_x_3 = lambda x, a:  x * 3 * a

numero_x_3(5,5)
```

¿por que utilizar funciones lambda? el poder de lambda se muestra cuando se utiliza como na funcion anonima dentr de otra funcion.

Digamos que tienes una definicion de funcion que toma un argunmento y ese argumento se multiplacara por un numero desconocido:

```python
#Ejemplo:
def myfunc(n):
    return lambda a : a * n

duplica =  myfunc(2)

print(duplica(11))


```

```python
suma_numeros = lambda num, num1 : num + num1
```

```python
print(suma_numeros(2, 4))
```
